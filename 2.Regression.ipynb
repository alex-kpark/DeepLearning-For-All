# 2.Multivariable Regression

If variable for the result goes more than 1(multivariable), We can simply increase the number of variable on hypothesis

But this form is not efficient, as the eqaution goes such long. Thus, we utilize **Matrix** to solve this issue. Utilizing **Matrix Multiplication**, we can solve the problem of lenghty Hypothesis.

### Hypothesis with formula way

```python
import tensorflow as tf

#data definition
x1_data = [73., 93., 89., 96., 73.,]
x2_data = [80., 88., 91., 98., 66.,]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142.]

#placeholders for variables, to figure out optimal value
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y= tf.placeholder(tf.float32)

#placeholders for weight
w1 = tf.Variable(tf.random_normal([1]), name='weight1') #variable with random value, but the number exists only one
w2 = tf.Variable(tf.random_normal([1]), name='weight2')
w3 = tf.Variable(tf.random_normal([1]), name='weight3')
b = tf.Variable(tf.random_normal([1]), name='bias')

#Hypothesis
hypothesis = x1*w1 + x2*w2 + x3*w3 + b
```

```python
#Cost Function and Gradient-Descent
cost = tf.reduce_mean(tf.square(hypothesis-Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
#Launch the Graph
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict={x1:x1_data, x2:x2_data, x3:x3_data, Y:y_data})
    if step % 10 == 0:
        print(step, "Cost ", cost_val, "\n Prediction:", hy_val)
```

### Matrix Way

```python
#data illustation in matrix

x_data = [[73., 80., 75.], [93., 88., 93.], [89.,91.,90.], [96., 98., 100.], [73., 66., 70.]] #5X3 Matrix
y_data = [[152.],[185.],[180.],[196.],[142.]] #5X1 Matrix

#placeholder for a tensor
#For placeholder, necessary to define number of instances
X = tf.placeholder(tf.float32, shape=[None, 3]) #shape=[Number of instances(None if not fixed), Number of variable]
Y = tf.placeholder(tf.float32, shape=[None, 1])

#For Variable, does not have to define the number of instances, just the form of matrix
W = tf.Variable(tf.random_normal([3,1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

#Hypothesis
hypothesis = tf.matmul(X,W) + b #matmul means matrix multiplication
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run(
    [cost, hypothesis, train], feed_dict={X:x_data, Y:y_data})
    
    if step % 10 == 0:
        print(step, cost_val, hy_val)
```

### Usage of Filename Queue & Batch

If we take a huge amout of data for the training, it would be hard a single memory can afford the whole proess. Thus, Tensorflow provides the concept of Queue, that generating the Queue of data and read it seperately, and decode it for example queue for actual training.

![queue](img/5.png)


1. It takes the array of filenames in string
2. It reads the key, and value
3. It records it with specified form of data (float maybe), and decode it as csv

```python
import tensorflow as tf

#step01: Queue the filename
filename_queue = tf.train.string_input_producer(['test_score.csv'], shuffle=False, name='filename_queue')

#step02: Reader, read the filename queue
reader = tf.TextLineReader()
key, value = reader.read(filename_queue)

#step03: define the type of data as float, and decode it
record_defaults = [[0.],[0.],[0.],[0.]]
xy = tf.decode_csv(value, record_defaults=record_defaults)

#Batch : Batch the data only with the size of 10 (batch size)
train_x_batch, train_y_batch = tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10)

#placeholder
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3,1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

#Hypothesis
hypothesis = tf.matmul(X,W) + b
cost = tf.reduce_mean(tf.square(hypothesis-Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

#Session
sess = tf.Session()
sess.run(tf.global_variables_initializer())

coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)

for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
    cost_val, hy_val, _ = sess.run([cost,hypothesis,train], feed_dict={X:x_batch, Y:y_batch})

    if step % 10 == 0: 
        print(step, cost_val, hy_val)
```
